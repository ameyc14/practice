### Above data is data for prebid auctions. There are bidders who receive requests for placing ads, they respond to these requests with a bid and the highest bid wins the auction. Analyze the dataset to answer which bidder has the highest win rate (wins / total_bid)?  

```python
import pandas as pd

df=pd.read_csv('analysis_id.csv')
```

```python
df.head()
```

```python
df=df[df['bidder'].notna()]
```

```python
df
```

```python
df['bidder']=df['bidder'].str.strip()
```

```python
df=df[df['bidder']!=""]
```

```python
df
```

```python
# grouped_data=df.groupby('bidder').agg(total_wins='sum_prebid_winning_bid', count)
grouped_data=df.groupby('bidder')['sum_prebid_winning_bid'].agg(total_wins='count',total_bids='size').reset_index()
grouped_data
```

```python
grouped_data['winrate']=grouped_data['total_wins']*100/grouped_data['total_bids']
grouped_data
```

```python
highest=grouped_data.loc[grouped_data['winrate'].idxmax()]
perc= grouped_data['winrate'].max()
```
