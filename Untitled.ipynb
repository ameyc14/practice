```python
import numpy as np
a=np.array([[30,40,50,30],[13,5,5,7]])
print(type(a))
arr=np.array(a)
print(type(arr))
print(arr)
print(arr.shape)
print(len(arr))
print(np.size(arr))
print(arr.astype(int)) # conversion of datatype
```

```python
a1=np.array([[30,40,50],[13,5,5]])
a2=([[30,40,50],[13,5,7]])
print(np.add(a1,a2))
print(np.subtract(a1,a2))
print(np.divide(a1,a2))
b=np.array_split(a1,3)
print(b)
print(np.insert(a1,1,[2,3,4], axis=1))
```
