```python
import pandas as pd

# DataFrame with test scores
data = {
    "Subject": ["Math", "Science", "English", "History"],
    "Mean_Score": [85, 78, 88, 92],
    "SD_Score": [5, 8, 6, 4]
}
scores = pd.DataFrame(data)

# Calculate CV for test scores
scores["CV"] = (scores["SD_Score"] / scores["Mean_Score"]) * 100

# Find the subject with the most consistent (min CV) and highest variability (max CV)
most_consistent = scores.loc[scores["CV"].idxmin()]
most_consistent
most_variable = scores.loc[scores["CV"].idxmax()]
most_variable
# # Print the insights
print(f"{most_consistent['Subject']} has the most consistent test scores (CV = {most_consistent['CV']:.2f}%).")
print(f"{most_variable['Subject']} has the highest variability (CV = {most_variable['CV']:.2f}%).")
```
